```python
import numpy as np
import pandas as pd
import os
import glob
import random
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
import scipy.stats as stats
```

### All data in /data/tvanasse/nccam3/data directory

```python
subdirs = []
for file in glob.glob('/data/tvanasse/nccam3/data/sub-2*'):
    subdirs.append(file)
subdirs.sort()
print("Subjects with cleaned data: " + str(len(subdirs)))

sesdirs = []
for i in range(0,len(subdirs)):
    for file in glob.glob(subdirs[i] + '/eeg/*'):
        sesdirs.append(file)
print("Nights cleaned: " + str(len(sesdirs)))
```

### Load dataframe

```python
nrem_dataframe = pd.read_csv('nrem_dataframe.csv')
```

### Create histogram of awakenings

```python
for i in range(0, len(nrem_dataframe)):
    if int(nrem_dataframe['Date_Time'].tolist()[i][11:13]) < 10:
           nrem_dataframe.loc[i,"Hour_Awoken"] = (int(nrem_dataframe['Date_Time'].tolist()[i][11:13]) + 24)
    else:
        nrem_dataframe.loc[i,"Hour_Awoken"] = int(nrem_dataframe['Date_Time'].tolist()[i][11:13])
```

```python
plt.figure(dpi = 300)

# REMOVE data with multiple dins
nrem_dataframe = nrem_dataframe[nrem_dataframe['DIN_EVENTS_WITHIN_TWO_MINUTES'] == 1].reset_index(drop=True)

from matplotlib.ticker import PercentFormatter, MultipleLocator

nrem_dataframe = nrem_dataframe[nrem_dataframe["Q1Experience"] != "False"]

nrem_dataframe["Q1 Response"] = "Report, N=" + str(len(nrem_dataframe[nrem_dataframe["Q1Experience"] == "Report"]))
nrem_dataframe.loc[nrem_dataframe[nrem_dataframe["Q1Experience"] == "Something"].index,"Q1 Response"] = "Something, N=" +\
    str(len(nrem_dataframe[nrem_dataframe["Q1Experience"] == "Something"]))
nrem_dataframe.loc[nrem_dataframe[nrem_dataframe["Q1Experience"] == "NoReport"].index,"Q1 Response"] = "No Report, N=" +\
    str(len(nrem_dataframe[nrem_dataframe["Q1Experience"] == "NoReport"]))


g = sns.displot(data=nrem_dataframe, 
                x="Hour_Awoken", 
                hue="Q1 Response", 
                discrete=True,
                multiple="stack")


ax = plt.gca()

ax.set_xticklabels(['20','21','22','23','0','1','2','3','4','5','6','7','8'])

for ax in g.axes.flat:
    ax.xaxis.labelpad = 8
    ax.yaxis.labelpad = 8
    ax.tick_params(which='both', width=0.8, labelsize=8)
    ax.xaxis.set_major_locator(MultipleLocator(1)) # x ticks at multiples of 1
#     ax.yaxis.set_major_formatter(PercentFormatter(1)) # percentage using 1 for 100%
#     ax.set_ylabel('Percentage')
    for p in ax.patches:
        h = p.get_height()
#         print(h)
        if h > 5: # skip empty bars
            txt = str(np.round(h)).split('.')[0]
#             print(txt)
            txt_x = p.get_x() + p.get_width() / 2
            txt_y = p.get_y() + h / 2
            ax.text(txt_x, txt_y, txt, ha='center', va='center')

            plt.subplots_adjust(left=0.09) # make a bit more room for the label

# for legend title
# g.legend(fontsize=15)

plt.ylabel("Awakenings Count")
plt.xlabel("Hour Awoken")
plt.title("Awakening Responses, N=" + str(len(nrem_dataframe)) + " Trials")
plt.show()


g.savefig('./figures/q1histogram_over_time.png',dpi=300)



```

```python
# remove data with multiple dins
nrem_dataframe = nrem_dataframe[nrem_dataframe['DIN_EVENTS_WITHIN_TWO_MINUTES'] == 1].reset_index(drop=True)

for min_experiments in [1,2,3]:

    sub_report_only = []
    sub_noreport_only = []
    sub_somethingandnoreport = []
    sub_reportorsomethingandnoreport = []
    sub_reportandnoreport = []
    sub_reportandsomething = []

    # loop through subjects
    for i in nrem_dataframe["ID"].unique():
    #     print(i)
    #     print(nrem_dataframe[(nrem_dataframe["ID"] == i)]["Q1Experience"].values)
        if (len(nrem_dataframe[(nrem_dataframe["ID"] == i) &\
                          (nrem_dataframe["Q1Experience"] == "Report")])) ==\
            (len(nrem_dataframe[(nrem_dataframe["ID"] == i)])):
            sub_report_only.append(i)

        if (len(nrem_dataframe[(nrem_dataframe["ID"] == i) &\
                          (nrem_dataframe["Q1Experience"] == "NoReport")])) ==\
            (len(nrem_dataframe[(nrem_dataframe["ID"] == i)])):
            sub_noreport_only.append(i)

        if (len(nrem_dataframe[(nrem_dataframe["ID"] == i) &\
                          (nrem_dataframe["Q1Experience"] == "Something")]) >= min_experiments) &\
           (len(nrem_dataframe[(nrem_dataframe["ID"] == i) &\
                          (nrem_dataframe["Q1Experience"] == "NoReport")]) >= min_experiments):
            sub_somethingandnoreport.append(i)

        if (len(nrem_dataframe[(nrem_dataframe["ID"] == i) &\
                          (nrem_dataframe["Q1Experience"] != "NoReport")]) >= min_experiments) &\
           (len(nrem_dataframe[(nrem_dataframe["ID"] == i) &\
                          (nrem_dataframe["Q1Experience"] == "NoReport")]) >= min_experiments):
    #         print("reportorsomethingandnoreport")
            sub_reportorsomethingandnoreport.append(i)

        if (len(nrem_dataframe[(nrem_dataframe["ID"] == i) &\
                          (nrem_dataframe["Q1Experience"] == "Report")]) >= min_experiments) &\
           (len(nrem_dataframe[(nrem_dataframe["ID"] == i) &\
                          (nrem_dataframe["Q1Experience"] == "NoReport")]) >= min_experiments):
    #         print("reportandnoreport")
            sub_reportandnoreport.append(i)

        if (len(nrem_dataframe[(nrem_dataframe["ID"] == i) &\
                                  (nrem_dataframe["Q1Experience"] == "Something")]) >= min_experiments) &\
            (len(nrem_dataframe[(nrem_dataframe["ID"] == i) &\
                                  (nrem_dataframe["Q1Experience"] == "Report")]) >= min_experiments):
                sub_reportandsomething.append(i)

    fig = plt.figure(dpi=400,figsize=(4,8))
#     fig, axs = plt.subplots(1, 1, )
    ax = fig.add_axes([0.1,0.4,0.75,0.60]) # axis starts at 0.1, 0.1
    langs = ['Subjects','w/ Report & NoReport', 
             'w/ Something & NoReport', 'w/ ReportORSomething &\nNoReport']
    students = [len(nrem_dataframe["ID"].unique()),len(sub_reportandnoreport), 
                len(sub_somethingandnoreport), len(sub_reportorsomethingandnoreport)]
    ax.bar(langs,students)
    plt.xticks(rotation=45)
    plt.ylabel('Subjects')
    plt.grid()
    
    plt.title('Distribution w/ ' + str(min_experiments) + ' Awakenings/Group/Subject')
    
    plt.savefig('./figures/q1histogram' + str(min_experiments) + '.png',dpi=300)
    plt.show()

    np.save('./sub_q1_data/sub_report_only',sub_report_only)
    np.save('./sub_q1_data/sub_noreport',sub_noreport_only)
    np.save('./sub_q1_data/sub_somethingandnoreport_min' + str(min_experiments),sub_somethingandnoreport)
    np.save('./sub_q1_data/sub_reportorsomethingandnoreport_min' + str(min_experiments),sub_reportorsomethingandnoreport)
    np.save('./sub_q1_data/sub_reportandnoreport_min' + str(min_experiments),sub_reportandnoreport)
    np.save('./sub_q1_data/sub_reportandsomething_min' + str(min_experiments),sub_reportandsomething)
```

## Out-of-Sample Data

```python
base = '/data/tvanasse/safmri/safmri_bids/'

first = 1
## find subject directories
dirs = os.listdir(base)
for x in range(0,len(dirs)):
#     print(dirs[x])
    for y in glob.glob(base + dirs[x] + '/ses-eeg1/eeg/*events.tsv'):
#         print(y)
        data = pd.read_csv(y, sep='\t')
        data["sub"] = dirs[x]

        if first == 1:
            veeg1_scores = data
            first = 0
        else:
            veeg1_scores = veeg1_scores.append(data, ignore_index=True)
            
## TRIGGER CODE (S  1)
veeg1_scores = veeg1_scores[veeg1_scores["value"] == 'S  1']

## CORRUPTED FILES
veeg1_scores = veeg1_scores[veeg1_scores['awak_number'].notna()].reset_index(drop=True)
veeg1_scores = veeg1_scores.drop(veeg1_scores[(veeg1_scores['sub'] == "sub-08") & (veeg1_scores['awak_number'] == 12)].index)
veeg1_scores = veeg1_scores.drop(veeg1_scores[(veeg1_scores['sub'] == "sub-08") & (veeg1_scores['awak_number'] == 13)].index)

veeg1_scores = veeg1_scores.reset_index(drop=True)

veeg1_scores["cleaned_path"] = ""

for i in range(0,len(veeg1_scores)):

    try:
        for file in os.listdir(base + "derivatives/eeglab/" + veeg1_scores["sub"][i] + '/ses-eeg1/eeg/'):
            if file.endswith("awak-" + (str(int(veeg1_scores["awak_number"][i]))) +\
                             "_hp_nbc_amica1_bsr_amica2_sc.set"):
                print(os.path.join(base + "derivates/eeglab/" + veeg1_scores["sub"][i] +\
                                   '/ses-eeg1/eeg/', file))
                veeg1_scores.loc[i,"cleaned_path"] = str(os.path.join(base + "derivatives/eeglab/" + veeg1_scores["sub"][i] +\
                                   '/ses-eeg1/eeg/', file))
    
    except FileNotFoundError:
        print("couldn't find: " + veeg1_scores["sub"][i])
    except:
        print("other error")
        
```

```python
veeg1_scores.to_csv('safmri_dataframe.csv', index=False)
```
